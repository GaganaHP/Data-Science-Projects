### Data Cleaning & Feature Engineering

```python
# Step 1: Combine Date and Time into a single column
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Step 2: Extract new features
df['Date'] = df['Datetime'].dt.date
df['Time'] = df['Datetime'].dt.time
df['Hour'] = df['Datetime'].dt.hour
df['DayOfWeek'] = df['Datetime'].dt.day_name()

# Step 3: Categorize into time intervals
def get_time_interval(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    else:
        return 'Evening'

df['Time_Interval'] = df['Hour'].apply(get_time_interval)
```

```python
# Drop any rows with missing values if they exist
df.dropna(inplace=True)

# Drop duplicates if any
df.drop_duplicates(inplace=True)
```
