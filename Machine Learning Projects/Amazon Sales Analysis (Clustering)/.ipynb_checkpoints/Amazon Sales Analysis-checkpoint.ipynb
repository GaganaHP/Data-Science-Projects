# Problem Statement: Clustering Products for Market Segmentation

## Objective:

Use unsupervised learning to cluster products based on their features, such as category, discounted_price, actual_price, discount_percentage, rating, rating_count, and about_product, to identify distinct product groupings or market segments.

## Purpose: 

Help Amazon or sellers discover patterns in product offerings, such as niche markets, high-demand product types, or oversaturated categories, to inform marketing, inventory management, or product development strategies.

## Applications:

* Identify groups of products with similar pricing and rating profiles for targeted promotions.
    
* Detect underserved product categories for new product launches.
                                       
* Understand competitive landscapes by grouping similar products (e.g., charging cables vs. kitchen appliances).

## Dataset Features for Clustering

The dataset provides a mix of numerical, categorical, and textual features, which can be used to create meaningful clusters. Here’s how each feature can contribute:

**Numerical Features:**

discounted_price: Reflects the selling price, indicating affordability or market positioning.

actual_price: Indicates the original price, useful for understanding premium vs. budget products.

discount_percentage: Captures pricing strategy (e.g., high discounts for clearance items).

rating: Measures customer satisfaction, indicating product quality or popularity.

rating_count: Reflects product popularity or market reach.

**Categorical Features:**

category: Groups products by type (e.g., “Computers&Accessories|Accessories&Peripherals” or “Home&Kitchen|Kitchen&HomeAppliances”). Can be encoded (e.g., one-hot encoding or label encoding) or used to create hierarchical clusters.

**Textual Feature:**

about_product: Product descriptions can be converted into numerical representations (e.g., TF-IDF vectors or Sentence-BERT embeddings) to capture semantic similarities (e.g., “fast charging” vs. “durable design”).

**Target:** 

None (unsupervised learning; the goal is to group products into clusters based on feature similarity).

**Encode Categorical Feature**

```python
from sklearn.preprocessing import OneHotEncoder

# Reset index to ensure alignment
df = df.reset_index(drop=True)

# Extract main_category
df['main_category'] = df['category'].str.split('|').str[0]

# One-hot encode main_category
encoder = OneHotEncoder(sparse_output=False, max_categories=10, handle_unknown='infrequent_if_exist')
category_encoded = encoder.fit_transform(df[['main_category']])
category_df = pd.DataFrame(category_encoded, columns=encoder.get_feature_names_out(['main_category']))

# Reset index of category_df to match df
category_df = category_df.reset_index(drop=True)

# Concatenate with df
df = pd.concat([df, category_df], axis=1)
```

### 5. Dimensionality Reduction with PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Apply PCA to reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Print explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Explained Variance:", sum(pca.explained_variance_ratio_))

# Visualize PCA results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], color='teal')
plt.title('PCA: 2D Projection of Products')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

### 6. Clustering with K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Elbow curve and Silhouette Score
inertia = []
silhouette_scores = []
K = list(range(2, 10))

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)  # Ensure X is normalized
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Plot elbow curve and Silhouette Score
plt.figure(figsize=(12, 5))

# Elbow plot
plt.subplot(1, 2, 1)
plt.plot(K, inertia, 'bo-', color='teal')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')

# Silhouette plot
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'ro-', color='coral')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```

**Insights:**

* The “elbow” (where the rate of decrease in inertia slows down) appears around k=4.

* The Silhouette Score peaks at k=3 with a value of 0.60, indicating the best balance of cluster cohesion and separation.

* The score at k=4 (0.58) is very close to the peak, suggesting that k=4 is also a good choice.

* After k=4, the score drops steadily, reaching 0.35 by k=9, indicating that adding more clusters leads to less distinct and more overlapping clusters.

* A Silhouette Score of 0.60 (at k=3) is decent, suggesting moderate to good cluster quality (typically, >0.5 is good, 0.3–0.5 is moderate, and <0.3 is poor). However, scores below 0.5 after k=5 suggest that higher k values may not yield meaningful clusters.

* Both metrics point to k=3 or k=4 as strong candidates.

* Since the Silhouette Score is slightly higher at k=3 (0.60 vs. 0.58), and a higher Silhouette Score indicates better-defined clusters, we’ll choose k=3 as the optimal number of clusters. However, k=4 could also be explored if we want more granular segments, given the elbow at k=4.

#### 6.1 Apply K-Means with k=3

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt

# Apply K-Means with optimal k=3
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Visualize clusters in PCA space
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['cluster'], palette='deep')
plt.title('K-Means Clusters (k=3) in PCA Space')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

**Insights:**

* Since the 2D PCA projection captures only 53.8% of the variance, the clusters might not appear well-separated in this plot, even if they are distinct in the original X space.

### 7. Evaluation and Interpretation

```python
print("\nCluster Sizes:")
print(df['cluster'].value_counts())
print("\nSilhouette Score for k={}:".format(optimal_k), silhouette_score(X, df['cluster']))
```

```python
# Cluster statistics
cluster_stats = df.groupby('cluster')[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].mean()
print("\nCluster Statistics:")
print(cluster_stats)

# Sample products per cluster (using main_category as proxy for product name)
for cluster in df['cluster'].unique():
    print(f"\nCluster {cluster} Sample Products:")
    print(df[df['cluster'] == cluster][['main_category', 'discounted_price', 'rating']].head())
```
