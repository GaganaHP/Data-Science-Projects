### OBJECTIVE

To build a Linear Regression model that accurately predicts the price of a diamond based on its various physical characteristics such as carat weight, cut, color, clarity, depth, table, and dimensions (x, y, z).

### PROBLEM STATEMENT

In the diamond industry, accurately pricing a diamond is crucial for maintaining competitiveness and customer trust.
Using the given dataset of diamond attributes and prices, the goal is to perform thorough data analysis and model building to develop a predictive system that can estimate the price of a diamond based on its features.
The project involves extensive Exploratory Data Analysis (EDA), feature engineering, statistical modeling using OLS, machine learning using Linear Regression, and model tuning to enhance prediction performance.

#### 6.2. Outlier Removal (Optional based on scatterplots)

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
for col in num_cols:
    if col != 'price':
        sns.scatterplot(x=col, y='price', data=df)
        plt.title(f'{col} vs price')
        plt.show()
```

### 7. Feature Engineering

```python
df1 = df.copy(deep = True)
```

```python
df1['volume'] = df1['x'] * df1['y'] * df1['z']
df1 = df1.drop(columns=['x', 'y', 'z'])
```

```python
sns.heatmap(df1[df1.describe().columns].corr(),annot=True)
```

### 8. Model Building

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

```python
df1.columns
```

```python
model = smf.ols(formula = "price ~ carat + cut + color + clarity + depth + table + volume", data = df1).fit()
model.summary()
```

```python
model1 = smf.ols(formula = "price ~ carat + cut + color + clarity + depth + table", data = df1).fit()
model1.summary()
```

```python
df1 = pd.get_dummies(df1,drop_first=True).astype('int')
df1
```

```python
df2 = df1.copy(deep=True)
```

```python
X = df2.drop('price', axis = 1)
y = df2.price
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 0)
```

```python
model3 = LinearRegression()
model3.fit(X_train, y_train)
```

```python
y_train_pred = model3.predict(X_train)
y_test_pred = model3.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```

```python
print(mean_squared_error(y_train,y_train_pred))
print(mean_squared_error(y_test,y_test_pred))
```

```python
print(np.sqrt(mean_squared_error(y_train,y_train_pred)))
print(np.sqrt(mean_squared_error(y_test,y_test_pred)))
```

```python
df2 = df2.drop(columns=['volume'])
```

```python
model4 = LinearRegression()
model4.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print('Train R2 Score:', r2_score(y_train, y_train_pred))
print('Test R2 Score:', r2_score(y_test, y_test_pred))
```
